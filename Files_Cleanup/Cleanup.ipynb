```python
# Dependencies and declarations
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
# Read the csv file
file_music = ("data.csv")
raw_music_data_df = pd.read_csv(file_music, delimiter ='#')
raw_music_data_df.head()
```

```python
# Add empty "Master Genre" column to be able to roll up all the smaller genre breakdowns
raw_music_data_df["Master Genre"] = ""
raw_music_data_df.head()
```

```python
# Populate the "Master Genre" column based on data in the genre_key.csv
# Open the genre_key.csv
genre_key = pd.read_csv("genre_key.csv")
music_data_df = pd.merge(raw_music_data_df, genre_key, how="left", on = ["Genre"])
music_data_df.drop('Master Genre_x', axis=1, inplace=True) 
music_data_df.rename(columns={"Master Genre_y":"Master Genre"}, inplace = True)
music_data_df.dtypes

```

```python
# Delimit the date with year and month
music_data_df["Date"] = pd.to_datetime(music_data_df["Date"])
# music_data_df.dtypes
#air_quality["month"] = air_quality["datetime"].dt.month
music_data_df["Year"] = music_data_df["Date"].dt.year
music_data_df["Month"] = music_data_df["Date"].dt.month
music_data_df
```
