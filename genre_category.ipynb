```python
#read the csv
# Dependencies
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
# Store filepath in a variable
file_music = Path("Music_Data.csv")
music_data_df = pd.read_csv(file_music, encoding="ISO-8859-1")
```

```python
# Plot data Genre by Year
# Genre by stream
# Genre 

summary_agg = music_data_df.groupby("Master Genre").agg({"Streams":["mean","median","var","std","sem"]})
summary_agg
```

```python
#Genre by Year
fig = plt.figure(figsize = (15, 6))
plt.bar(music_data_df["Year"],music_data_df["Master Genre"], align ="center")
plt.title('Total Number of Streams by Year')
plt.xlabel("Streams")
plt.ylabel('Year')
plt.xticks(rotation=90)
plt.show()
```

```python
#Streams by Genre
fig = plt.figure(figsize = (10, 6))
plt.bar(music_data_df["Master Genre"],music_data_df["Streams"], align ="center")
plt.title('Total Number of Streams by Genre')
plt.xlabel("Streams")
plt.ylabel('Year')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 5 artist by year
fig = plt.figure(figsize = (10, 6))
plt.bar(music_data_df["Artist"],music_data_df["Year"], align ="center")
plt.title('Total Number of Streams by ')
plt.xlabel("Streams")
plt.ylabel('Year')
plt.xticks(rotation=90)
plt.show()
```
