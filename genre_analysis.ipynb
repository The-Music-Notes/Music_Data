```python
#read the csv
# Dependencies
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
# Store filepath in a variable
file_music = Path("Files_Cleanup/Music_Data.csv")
music_data_df = pd.read_csv(file_music, encoding="ISO-8859-1")
```

```python
# Show the first five rows.
music_data_df.head()
```

```python
# Plot data Genre Summary by Streams

summary_agg = music_data_df.groupby("Master Genre").agg({"Streams":["mean","median","var","std","sem"]})
summary_agg
```

```python
#Split the master data by Year

summary_year_2017 = pd.DataFrame(music_data_df[music_data_df['Date'].str[-4:]=='2017'])
summary_year_2018 = pd.DataFrame(music_data_df.loc[music_data_df['Date'].str[-4:]=='2018'])
summary_year_2019 = pd.DataFrame(music_data_df.loc[music_data_df['Date'].str[-4:]=='2019'])
summary_year_2020 = pd.DataFrame(music_data_df.loc[music_data_df['Date'].str[-4:]=='2020'])
summary_year_2021 = pd.DataFrame(music_data_df.loc[music_data_df['Date'].str[-4:]=='2021'])

summary_agg_year_2017 = summary_year_2017.groupby(["Master Genre", "Year"  ])["Streams"].aggregate('count')

summary_agg_year_2017 
# Plot aggreagate for various years 2017 - 2021 using above data filtered by year

fig = plt.figure(figsize = (8,5))
summary_agg_year_2017.plot(kind='bar', color = 'red')

plt.title('Total Number of Streams by Genre and Year = 2017')
plt.xlabel("Genre")
plt.ylabel('Streams')
plt.xticks(rotation=90)
plt.show()

#2017
```

```python
#Streams by Genre for all years 2017-2021. Streams are in large numbers and scaled down

fig = plt.figure(figsize = (10, 6))
plt.bar(music_data_df["Master Genre"],music_data_df["Streams"], align ="center")
plt.title('Total Number of Streams by Genre - All years')
plt.xlabel("Genre")
plt.ylabel('Streams')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top Genre in all years

color = ['lightblue', 'blue', 'purple', 'red', 'black','yellow', 'orange','pink']

summary_agg_genre = music_data_df['Master Genre'].value_counts()
summary_agg_genre.plot(kind='bar', color = color)
plt.title('Top Genre - all years')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
#Filter by Single artist and show streams for all years

summary_artist = pd.DataFrame(music_data_df[music_data_df['Artist'] == 'Taylor Swift'])
summary_artist_2017 = pd.DataFrame(summary_artist[summary_artist['Date'].str[-4:]=='2017'])
summary_artist_2018 = pd.DataFrame(summary_artist[summary_artist['Date'].str[-4:]=='2018'])
summary_artist_2019 = pd.DataFrame(summary_artist[summary_artist['Date'].str[-4:]=='2019'])
summary_artist_2020 = pd.DataFrame(summary_artist[summary_artist['Date'].str[-4:]=='2020'])
summary_artist_2021 = pd.DataFrame(summary_artist[summary_artist['Date'].str[-4:]=='2021'])
summary_artist_2021
```

```python
#Create Line plot for artist

summary_artist.groupby('Year')['Streams'].sum().plot(legend=True, color = 'red')
plt.xlabel('Year')
plt.ylabel('Taylor Swift Streams')
```

```python
taylor_bar=summary_artist.groupby('Year')['Streams'].sum()
plt.xlabel('Year')
plt.ylabel('Streams')
taylor_bar.plot(kind='bar', color = 'green')
```

```python
#Get top 5 artists for all years
summary_agg_artist = pd.DataFrame(music_data_df.groupby(['Artist'])['Streams'].aggregate('count'))

summary_agg_artist.sort_values(by=['Streams'], ascending=False)
```

```python
#Top 5 artists

summary_agg_artist_new = pd.DataFrame(summary_agg_artist.loc[summary_agg_artist['Streams'] >= 6496])
summary_agg_artist_new.sort_values(by=['Streams'], ascending=False)
summary_agg_artist_new
```

```python
summary_agg_artist_new_index = summary_agg_artist_new.copy()
summary_agg_artist_new_index.reset_index(inplace=True)
summary_agg_artist_new_index
```

```python
#Percentage of artists streams across all years
labels = summary_agg_artist_new_index['Artist']


colors = ['Orange','Blue','Green','Red','Pink']
explode = (0,0,0.05,0,0)
x=summary_agg_artist_new_index['Streams']
plt.pie(x=x, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=False)
plt.show()
```
